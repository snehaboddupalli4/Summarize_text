```python
import requests
import PyPDF2
import docx2txt
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lex_rank import LexRankSummarizer

def is_drive_link_accessible(drive_link):
    response = requests.head(drive_link)
    return response.status_code == 200

while True:
    print("Select the type of file to be summarized:")
    print("1. PDF")
    print("2. Text file")
    print("3. Direct text")
    print("q. Quit")
    choice = input("Enter your choice: ")

    if choice == "q":
        break

    if choice == "1":
        drive_link = input("Enter the Google Drive link to the PDF file: ")

        if not is_drive_link_accessible(drive_link):
            print("Unable to access the file.")
            continue

        # Convert to direct download link
        file_id = drive_link.split("/")[-2]
        direct_download_link = f"https://drive.google.com/uc?id={file_id}"

        # Download the PDF file
        response = requests.get(direct_download_link)
        with open("temp.pdf", 'wb') as file:
            file.write(response.content)

        # Extract text from PDF
        with open("temp.pdf", 'rb') as file:
            pdf_reader = PyPDF2.PdfReader(file)
            text = ""
            for page in pdf_reader.pages:
                text += page.extract_text()

    elif choice == "2":
        drive_link = input("Enter the Google Drive link to the text file: ")

        if not is_drive_link_accessible(drive_link):
            print("Unable to access the file.")
            continue

        # Convert to direct download link
        file_id = drive_link.split("/")[-2]
        direct_download_link = f"https://drive.google.com/uc?id={file_id}"

        # Download the text file
        response = requests.get(direct_download_link)
        with open("temp.txt", 'wb') as file:
            file.write(response.content)

        # Read text from the file
        with open("temp.txt", 'r') as file:
            text = file.read()

    elif choice == "3":
        text = input("Enter the text: ")

    else:
        print("Invalid choice.")
        continue

    # Summarize the text
    parser = PlaintextParser.from_string(text, Tokenizer("english"))
    summarizer = LexRankSummarizer()
    summary = summarizer(parser.document, sentences_count=3)

    # Display the summary
    print("Summary:")
    for sentence in summary:
        print(sentence)

    print("---------------------")
```
