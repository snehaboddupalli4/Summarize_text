```python
import telebot
from transformers import pipeline

# Set up Telegram Bot API token
bot_token = '6210741397:AAFihcKH5tThTjR32LRaUiqXrZoWg8eUveI'
bot = telebot.TeleBot(bot_token)

# Set up summarization pipeline
summarizer = pipeline("summarization")

# Handler for /start and /help commands
@bot.message_handler(commands=['start', 'help'])
def send_welcome(message):
    bot.reply_to(message, "Welcome! Send me a text, and I will summarize it for you.")

# Handler for incoming messages
@bot.message_handler(func=lambda message: True)
def summarize_text(message):
    text = message.text

    # Perform summarization
    summary = summarizer(text, max_length=100, min_length=30, do_sample=False)

    # Extract summarized text from the response
    summarized_text = summary[0]['summary_text']

    # Send the summarized text as a reply
    bot.reply_to(message, summarized_text)

# Start the bot
bot.polling()
```
